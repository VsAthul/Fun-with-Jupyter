```python
import sqlite3
```

```python
conn = sqlite3.connect('test_db')
```

```python
cursor_obj= conn.cursor()
```

```python
table=""" create table instructor(id int primary key not null, fname varchar(20), lname varchar(20), city varchar(10), ccode char(2))"""
cursor_obj.execute(table)
print("Table is ready")
```

```python
cursor_obj.execute(''' insert into instructor values(1, 'Rav', 'Ahuja', 'TORONTO', 'CA'), (2, 'Raul', 'Chong', 'Markham', 'CA'), (3, 'Hima', 'Vasudevan', 'Chicago', 'US') ''')
```

```python
statement=""" select * from instructor """
cursor_obj.execute(statement)
print("All the data")
outputall=cursor_obj.fetchall()
print(outputall)
for row in outputall:
  print(row)
```
